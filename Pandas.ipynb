## **What to do when we have NaN values?**
Count missing values per column:

```python
data3 = pd.read_csv('train.csv')
```

```python
data3.isna().sum()
```

```python
# drop missing rows or cols: (only if there are not a lot of missing values)
# drop rows:
# data3.dropna()
# drop cols:
# data3.dropna(axis=1)
```

```python
# fill missing values with a constant
cols = data3.columns[data3.isna().any()].tolist()
for col in cols:
  data3[col].fillna(0, inplace=True)
data3.isna().sum()
```

```python
## fill with the mean (numbers)
data4 = pd.read_csv('train.csv')
#data4.isna().sum()
cols = data4.columns[data4.isna().any()].tolist()
if cols:
    data4[cols[0]] = data4[cols[0]].fillna(data4[cols[0]].mean())
data4.isna().sum()
```

```python
len(data4.Cabin.unique())
```

```python
data4[cols[1]] = data4[cols[1]].fillna(data4[cols[1]].mode()[0])
data4.isna().sum()
```

```python
data4 = data4.dropna()
data4.isna().sum()
```
